```python
import pandas as pd
import numpy as np
```

Loading IAS Data from Different Sheets of an Excel File

```python

data = pd.DataFrame()

data

for i in range(1, 26):
    temp =  pd.read_excel('IAS.xlsx', sheet_name = f'Sheet{i}')
    data = pd.concat([data, temp], axis=0)

data
```

Extracting Name, ID, Gender, Date of Appointment, Source Of Recruitment, Date Of Birth, Allotment Year, Present Post, Current Post, Cadre and Pay Level

```python

name = data.iloc[:, 1].apply(lambda x: x.split('_x000D_')[0])
id = data.iloc[:, 1].apply(lambda x: x.split('_x000D_')[1].strip().split('No. ')[1].strip())
gender = name.apply(lambda x: 'Male' if 'Shri' in x else 'Female')

date_of_appointment = data.iloc[:, 1].apply(lambda x: pd.to_datetime(x.split('_x000D_')[2].strip().split(';')[0]).date())
source_of_recruitment = data.iloc[:, 1].apply(lambda x: x.split('_x000D_')[2].strip().split(';')[1])

dob = data.iloc[:, 3].apply(lambda x: pd.to_datetime(x.split('_x000D_')[0]).date())
allotment_year = data.iloc[:, 3].apply(lambda x: x.split('_x000D_')[1].strip().split(';')[0])

present_post = data.iloc[:, 4].apply(lambda x: x.split('_x000D_')[0])
current_post = present_post.apply(lambda x: x.split(', ')[0])

cadre = data.iloc[:, 3].apply(lambda x: x.split('_x000D_')[1].strip().split(';')[1])
pay_level = data.iloc[:, 5].apply(lambda x: ' '.join (x.split('_x000D_')[0].split(' ')[:2]))

```

Extracting With Effect From Date

```python

def get_wef_date(x):
    x = x.split('_x000D_')
    if len(x) == 2:
        return pd.to_datetime(x[1].strip()).date()
    else:
        return np.nan

wef_date = data.iloc[:, 4].apply(get_wef_date)
```

Extracting Domicile, Department Name and Location

```python

def get_domicile(x):
    x = x.split('_x000D_')
    if len(x) == 3:  return x[2].strip()
    else:  return np.nan


def get_dept(x):
    x = x.split(', ')
    if len(x) > 2:  return x[1].strip()
    return np.nan


def get_loc(x):
    x = x.split(', ')
    if len(x) > 1:  return x[-1].strip()
    return np.nan


domicile = data.iloc[:, 3].apply(get_domicile)
dept = present_post.apply(get_dept)
location = present_post.apply(get_loc).to_list()
```

Extracting Edu. Qualification

```python

def get_qual2(x, ind = 2):

    if type(x) != float:
        x = x.split('_x000D_')
        if len(x) > 1:  return x[1].strip()
    return np.nan


def get_specialization(x):

    if type(x) == float:  return np.nan

    x = x.split('(')
    if len(x) > 1:  return x[1].strip().replace(')', '')
    return np.nan


qual1 = data.iloc[:, 2].apply(lambda x: x.split('_x000D_')[0].strip() if type(x) != float else np.nan)
qual2 = data.iloc[:, 2].apply(get_qual2)

spec1 = qual1.apply(get_specialization)
spec2 = qual2.apply(get_specialization)

qual1 = qual1.apply(lambda x: x.split('_x000D_')[0].split('(')[0].strip() if type(x) != float else np.nan)
qual2 = qual2.apply(lambda x: x.strip().split('(')[0] if type(x) != float else np.nan)
```

Assigning Extracted Data to Pandas Dataframe

```python
final_df = pd.DataFrame()

final_df['ID'] = id
final_df['Name'] = name
final_df['Gender'] = gender
final_df['DOB'] = dob


final_df['Date of Appointment'] = date_of_appointment

final_df['Source of Recruitment'] = source_of_recruitment
final_df['Allotment Year'] = allotment_year
final_df['Cadre'] = cadre
final_df['Qual 1'] = qual1
final_df['Qual 2'] = qual2


final_df['Spec 1'] = spec1
final_df['Spec 2'] = spec2

final_df['Domicile'] = domicile
final_df['Current Post'] = current_post
final_df['Department'] = dept
final_df['Location'] = location

final_df['With Effect From Date'] = wef_date
final_df['Pay Level'] = pay_level

final_df.to_csv('IAS.csv', index=False)
final_df
```
